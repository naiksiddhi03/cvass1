<u>USING GAUSSIAN BLUR</u>  

```python
#Uploading image and reading it.

#from google.colab import files
import cv2
#uploaded = files.upload()
image = cv2.imread('E:\Btech sem 7\cv\scenery.jfif', -1)
```

```python
for i in range (4):
  filtersize = [3, 5, 11, 15]
  filtersizeTitle = ['3 X 3', '5 X 5', '11 X 11', '15 X 15']
  print(filtersizeTitle[i])
  gaussian = cv2.GaussianBlur(image, ksize = (filtersize[i], filtersize[i]), sigmaX = 0, sigmaY = 0)
  cv2.imshow(gaussian)
```
